<h1>import libraries</h1>

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

<h1>install folium, import HeatMap</h1>

```python
import folium
from folium.plugins import HeatMap
```

<h1>load dataframe</h1>

```python
gt = pd.read_csv('globalterro.csv', encoding='ISO-8859-1')
```

```python
gt.columns
```

```python
f = gt[gt['region_txt'] == 'Middle East & North Africa']
f = f[f['iyear'] >= 2000]
f = f[f['attacktype1_txt'] == 'Bombing/Explosion']
f = f[f['nkill'] >= 10]
print(f'Filtered dataset size: {len(f)} rows.')
```

```python
aggr = f.groupby(['latitude','longitude']).size().reset_index(name='incident_count')
print(aggr.head())
```

<h1>Visualizing geospatial using folium lib</h1>

```python
aggr = f.groupby(['latitude','longitude']).size().reset_index(name='incident_count')

aggr['normalized_weight'] = (aggr['incident_count'] - aggr['incident_count'].min()) / \
(aggr['incident_count'].max() - aggr['incident_count'].min())

heat_data = aggr[['latitude', 'longitude', 'normalized_weight']].values.tolist()

m = folium.Map(location=[30,40], zoom_start = 4)

HeatMap(heat_data).add_to(m)

# m.save('terro_heatmap.html')
m
```

<h1>regional analysis</h1>
<h3>By aggregating and mapping incident data specifically for South Asia from 2003 onwards, the resulting heatmap visually pinpoints geographical areas with the highest concentration of incidents, enabling targeted interventions and resource allocation to mitigate future occurrences in these identified hotspots.</h3>

```python
regional_analysis = gt[gt['region_txt'] == 'Middle East & North Africa']

regional_analysis = regional_analysis[regional_analysis['iyear'] >= 2003]

regional_analysis = regional_analysis[regional_analysis['attacktype1_txt']=='Hostage Taking (Kidnapping)']

regional_analysis = regional_analysis[regional_analysis['nkill'] >= 5]

print(f'filtered dataset: {len(regional_analysis)} rows.')
```

```python
aggregated_regional_analysis = regional_analysis.groupby(['latitude','longitude']).size().reset_index(name='incident_count')
print(aggregated_regional_analysis.head())
```

```python
regional_analysis = gt[(gt['region_txt'] == 'South Asia') & (gt['iyear'] >= 1999)]

aggregated_regional_analysis = regional_analysis.groupby(['latitude','longitude']).size().reset_index(name='incident_count')

aggregated_regional_analysis['normalized_weigh'] = (aggregated_regional_analysis['incident_count'] - aggregated_regional_analysis['incident_count'].min()) / \
(aggregated_regional_analysis['incident_count'].max() - aggregated_regional_analysis['incident_count'].min())

heat_data = aggregated_regional_analysis[['latitude','longitude','incident_count']].values.tolist()

regional_m = folium.Map(location=[30,40], zoom_start=4)

HeatMap(heat_data).add_to(regional_m)

regional_m
```

<h1>target analysis</h1>

```python
target_analysis = gt[gt['targtype1_txt'] == 'Private Citizens & Property']

target_analysis = target_analysis[target_analysis['targsubtype1_txt']=='Unnamed Civilian/Unspecified']

target_analysis = target_analysis[target_analysis['targsubtype1'] >= 10]

print(f'filtered dataset: {len(target_analysis)} rows.')
```

```python
aggregated_target_analysis = target_analysis.groupby(['latitude','longitude']).size().reset_index(name='incident_count')
print(aggregated_target_analysis.head())
```
