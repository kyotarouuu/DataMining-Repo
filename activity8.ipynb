<h1>Import libraries</h1>

```python
import numpy as np
import pandas as pd
import folium
from folium.plugins import HeatMap 
```

<h1>Load dataframe</h1>

```python
cc = pd.read_csv('chicagocrimes.csv')
cc
```

<hr>

<h1>Stalking Analysis 2025</h1>

```python
a = cc[(cc['Primary Type'] == 'STALKING')&(cc['Year']==2025)]

adf = a.groupby(['Latitude','Longitude']).size().reset_index(name='incident_count')

adf['normalized_weight'] = (adf['incident_count'] - adf['incident_count'].min()) / \
                            (adf['incident_count'].max() - adf['incident_count'].min())

heat_data = adf[['Latitude','Longitude','incident_count']].values.tolist()

amap = folium.Map(location=[41.8781, -87.6298], zoom_start=11)

HeatMap(heat_data).add_to(amap)

# amap.save('stalking.html')
amap
```
